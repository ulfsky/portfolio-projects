# Technical Requirements

Business goal: analyze trending-video history on YouTube <br>
How often the dashboard will be used: at least once a day <br>
Target dashboard user: video ads planning managers<br>

Dashboard data content:<br>
- Trending videos from the past, broken down by day and category<br>
- Likes, broken down by video category<br>
- Dislikes, broken down by video category <br>

Parameters according to which the data is to be grouped: <br>
- Trending date and time (don't forget that likes and dislikes must be calculated as of the last trending day of the video)<br>
- Video category <br>
- Region <br>

The nature of the data: <br>
- Trending history — absolute values with a breakdown by day <br> 
- Likes broken down by category — relative values (% of total likes)<br>
- Dislikes broken down by category — relative values (% of total dislikes)<br>

Importance: the trending-video history graph is the most important, so it should take up at least half of the dashboard <br>
Data sources for the dashboard: raw data on trending (the trending_vids table) <br>
Database storing aggregate data: additional aggregate tables in the youtube database<br>
Data update interval: once every 24 hours, at midnight UTC; <br>
Graphs to be displayed and their order, dashboard controls to be included<br>
